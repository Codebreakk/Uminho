```python
def fizzbuzz(n):
    for i in range(1,n+1):
        fizz = i%2
        buzz = i%5
        
        if fizz == 0 and buzz == 0:
            print("fizzbuzz")
        elif fizz == 0:
            print("fizz")
        elif buzz == 0:
            print("buzz")
        else:
            print(i)
            
fizzbuzz(100)
```
